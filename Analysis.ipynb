```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('epi_r_cleaned_smaller1.csv')
df.head(6)
```

```python
plt.figure(figsize=(10, 6))
df['rating'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='calories', y='fat', data=df, color='purple')
plt.title('Calories vs. Fat')
plt.xlabel('Calories')
plt.ylabel('Fat')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['calories'].plot(kind='hist', bins=20, color='green', alpha=0.7)
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
from collections import Counter

 
if {'calories', 'fat', 'protein'}.issubset(df.columns):
    nutrition_correlation = df[['calories', 'fat', 'protein', 'rating']].corr()
    print("Correlation between nutritional content and ratings:")
    print(nutrition_correlation)
    
   
    plt.figure(figsize=(8, 6))
    sns.heatmap(nutrition_correlation, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Correlation Between Nutritional Content and Ratings")
    plt.show()

    
    plt.figure(figsize=(10, 6))
    sns.histplot(df['calories'].dropna(), bins=20, color='green', kde=True)
    plt.title("Distribution of Calories in Recipes")
    plt.xlabel("Calories")
    plt.ylabel("Frequency")
    plt.show()
```
